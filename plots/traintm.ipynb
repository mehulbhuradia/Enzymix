```python
# get list of files in ./pdb
# for each file get name

import os

files = os.listdir('./generated_final/pdb')

tm_files = []
for file in files:
    if 'tm' in file and 'ref' not in file:
        tm_files.append(file)
```

```python
# load all tm files into a dictionary
import json
tm_dict = {}
for file in tm_files:
    with open('./generated_final/pdb/' + file) as f:
        data = json.load(f)
        tm_dict.update(data)
```

```python
len(tm_dict)
```

```python
# # Load all the values into a list
tm_scores_list = list(tm_dict.values())
tm_scores_list = [float(i) for i in tm_scores_list]

# sort the list
tm_scores_list.sort(key = float)

# print(tm_scores_list)

# plot the list
import matplotlib.pyplot as plt


plt.plot(tm_scores_list)
plt.ylabel('Training TM Score')

plt.show()
```

```python
# Make a histogram
plt.figure(figsize=(6, 4))
plt.hist(tm_scores_list, bins=10)
plt.ylabel('Frequency', fontsize=16)
plt.xlabel('Training TM Score', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.savefig('./saved_plots/traintm_hist.png', dpi=300, bbox_inches='tight')
plt.show()
```
