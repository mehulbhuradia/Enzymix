```python
# Load a dict from json file
import json
import matplotlib.pyplot as plt
```

```python
with open('sctm_scores.json', 'r') as f:
    tm_dict = json.load(f)

tm_scores_list = list(tm_dict.values())
tm_scores_list = [float(i) for i in tm_scores_list]
tm_scores_list.sort(key = float)
```

```python
# Make a histogram
# Add a line at 0.5 and label it as cutoff

plt.figure(figsize=(6, 4))
plt.hist(tm_scores_list, bins=10)
plt.axvline(x=0.5, color='r', linestyle='--', label='cutoff')
plt.legend()
plt.ylabel('Frequency', fontsize=16)
plt.xlabel('scTM Score', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)
# plt.title('scTM Scores of the generated structures')
plt.savefig('./saved_plots/sctm_hist.png', dpi=300, bbox_inches='tight')
plt.show()
```
